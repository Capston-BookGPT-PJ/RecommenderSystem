## 유사도 + 대출수 기반 가중 점수 계산

```python
'''# ======================================
# 5️⃣ 유사도 + 대출수 기반 가중 점수 계산 함수
# ======================================
def recommend_book(isbn=None, title=None, author=None, top_n=5, alpha=0.95):
    """
    📘 통합 추천 함수
    - ISBN이 있으면: 기존 책 데이터에서 추천
    - ISBN이 없고 title/author가 있으면: 새 임베딩으로 유사도 기반 추천

    alpha: 유사도(0.95) + 인기도(0.05) 가중 비율
    """
    # ==============================
    # ① ISBN 기반 추천
    # ==============================
    if isbn:
        isbn = str(isbn).strip()
        hits = df_books.index[df_books['ISBN_THIRTEEN_NO'] == isbn].tolist()
        if hits:
            idx = hits[0]
            qvec = embeddings[idx:idx+1]
        else:
            print(f"'{isbn}' 책을 데이터에서 찾을 수 없습니다. → 새 입력 기반 추천으로 전환합니다.")
            isbn = None  # fallback
    # ==============================
    # ② 새 입력 (제목/저자) 기반 추천
    # ==============================
    if not isbn:
        if not title:
            return "ISBN이 없을 경우, title은 반드시 입력해야 합니다."
        query_text = f"{title} {author or ''}".strip()
        qvec = model.encode([query_text], normalize_embeddings=True).astype('float32')

    # ==============================
    # ③ FAISS 검색
    # ==============================
    sims, inds = index.search(qvec, top_n + 1)
    rec_list = []
    for j, sim in zip(inds[0], sims[0]):
        # ISBN 기반인 경우 자기 자신 제외
        if isbn and j == idx:
            continue
        book = df_books.iloc[j]
        lon_score = book['LON_CO']
        lon_weight = np.log1p(lon_score) / np.log1p(df_books['LON_CO'].max())
        final_score = alpha * sim + (1 - alpha) * lon_weight

        rec_list.append({
            'ISBN_THIRTEEN_NO': book['ISBN_THIRTEEN_NO'],
            'BOOK_TITLE_NM': book['BOOK_TITLE_NM'],
            'AUTHR_NM': book['AUTHR_NM'],
            'PUBLISHER_NM': book['PUBLISHER_NM'],
            'KDC_NM': book['KDC_NM'],
            'LON_CO': lon_score,
            'cosine_similarity': round(float(sim), 4),
            'final_score': round(float(final_score), 4)
        })

    rec_df = pd.DataFrame(rec_list)
    rec_df.sort_values('final_score', ascending=False, inplace=True)
    return rec_df.head(top_n)'''
```

```python

def recommend_book(isbn=None, title=None, author=None, top_n=5, alpha=0.95):
    """
    📘 통합 추천 함수 (JSON 반환)
    - ISBN이 있으면: 기존 책 데이터에서 추천
    - ISBN이 없고 title/author가 있으면: 새 임베딩으로 유사도 기반 추천
    - 결과를 list[dict] 형태로 반환 (API 응답용)
    """
    # ==============================
    # ① ISBN 기반 추천
    # ==============================
    if isbn:
        isbn = str(isbn).strip()
        hits = df_books.index[df_books['ISBN_THIRTEEN_NO'] == isbn].tolist()
        if hits:
            idx = hits[0]
            qvec = embeddings[idx:idx+1]
        else:
            print(f"'{isbn}' 책을 데이터에서 찾을 수 없습니다. → 새 입력 기반 추천으로 전환합니다.")
            isbn = None  # fallback

    # ==============================
    # ② 새 입력 (제목/저자) 기반 추천
    # ==============================
    if not isbn:
        if not title:
            return {"error": "ISBN이 없을 경우, title은 반드시 입력해야 합니다."}
        query_text = f"{title} {author or ''}".strip()
        qvec = model.encode([query_text], normalize_embeddings=True).astype('float32')

    # ==============================
    # ③ FAISS 검색
    # ==============================
    sims, inds = index.search(qvec, top_n + 1)
    rec_list = []
    for j, sim in zip(inds[0], sims[0]):
        # ISBN 기반인 경우 자기 자신 제외
        if isbn and j == idx:
            continue

        book = df_books.iloc[j]
        lon_score = book['LON_CO']
        lon_weight = np.log1p(lon_score) / np.log1p(df_books['LON_CO'].max())
        final_score = alpha * sim + (1 - alpha) * lon_weight

        rec_list.append({
            'ISBN_THIRTEEN_NO': book['ISBN_THIRTEEN_NO'],
            'BOOK_TITLE_NM': book['BOOK_TITLE_NM'],
            'AUTHR_NM': book['AUTHR_NM'],
            'PUBLISHER_NM': book['PUBLISHER_NM'],
            'KDC_NM': book['KDC_NM'],
            'LON_CO': int(lon_score),
            'cosine_similarity': round(float(sim), 4),
            'final_score': round(float(final_score), 4)
        })

    rec_df = pd.DataFrame(rec_list)
    rec_df.sort_values('final_score', ascending=False, inplace=True)

    # ==============================
    # ④ 원본 데이터에서 이미지 등 추가
    # ==============================
    '''merged = pd.merge(
        rec_df,
        df_books_contents[['ISBN_THIRTEEN_NO', 'BOOK_IMAGE_NM', 'PBLICTE_DE']],
        on='ISBN_THIRTEEN_NO',
        how='left'
    )

    merged['BOOK_IMAGE_NM'] = merged['BOOK_IMAGE_NM'].replace('', 'No Image')'''


    # 추천 함수에서 merge 시 df_books_meta 사용
    merged = pd.merge(
        rec_df,
        df_books_meta,
        on='ISBN_THIRTEEN_NO',
        how='left'
    ).drop_duplicates(subset=['ISBN_THIRTEEN_NO'])

    merged['BOOK_IMAGE_NM'] = merged['BOOK_IMAGE_NM'].replace('', 'No Image')

    # ==============================
    # ⑤ JSON 형식 반환
    # ==============================
    result_json = merged.head(top_n).to_dict(orient='records')
    return result_json
```

```python
# ✅ 1️⃣ 기본 세팅
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd, numpy as np, re, faiss, torch, gc
from sentence_transformers import SentenceTransformer

path = '/content/drive/MyDrive/recommend_system/'

# ✅ 2️⃣ 데이터 로드
df_books_contents = pd.read_csv(path + 'NL_BO_BEST_LOAN_BOOK_HISTORY_202112.csv')

# ======================================
# 🔹 제목 정제 (제로폭/공백 제거)
# ======================================
def clean_title(x):
    if pd.isna(x): return ''
    return re.sub(r'[\s\u200b\xa0]+', ' ', str(x)).strip()

df_books_contents['BOOK_TITLE_NM'] = df_books_contents['BOOK_TITLE_NM'].apply(clean_title)
df_books = df_books_contents[df_books_contents['BOOK_TITLE_NM'] != '']

# ======================================
# 🔹 데이터 필터링 및 통합
# ======================================
df_books = df_books_contents[df_books_contents['ANALS_TY_CD_NM'] != '지역별'].copy()
target_ages = ['청소년(14~19)', '20대', '30대']
df_books = df_books[df_books['AGE_FLAG_NM'].isin(target_ages)].copy()

cols = [
    'ISBN_THIRTEEN_NO', 'BOOK_TITLE_NM', 'AUTHR_NM',
    'BOOK_INTRCN_CN', 'PUBLISHER_NM', 'KDC_NM', 'LON_CO'
]
df_books = df_books_contents[cols].copy().fillna('')

# ISBN, 제목, 저자 기준 통합 (대출수 합산)
df_books = (
    df_books.groupby(['BOOK_TITLE_NM', 'AUTHR_NM'], as_index=False)
    .agg({
        'ISBN_THIRTEEN_NO': 'first',
        'BOOK_INTRCN_CN': 'first',
        'PUBLISHER_NM': lambda x: ', '.join(set(x)),
        'KDC_NM': 'first',
        'LON_CO': 'sum'
    })
)

# ======================================
# 🔹 텍스트 피처 결합 (임베딩용)
# ======================================
df_books['text_for_embed'] = (
    df_books[['BOOK_TITLE_NM', 'AUTHR_NM', 'BOOK_INTRCN_CN', 'PUBLISHER_NM', 'KDC_NM']]
    .fillna('')
    .astype(str)
    .agg(' '.join, axis=1)
)

# ======================================
# 🔹 SentenceTransformer 임베딩
# ======================================
!pip install -q sentence-transformers faiss-cpu

device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = SentenceTransformer("jhgan/ko-sroberta-multitask", device=device)
model.max_seq_length = 384

texts = df_books['text_for_embed'].tolist()
torch.cuda.empty_cache(); gc.collect()

embeddings = model.encode(
    texts,
    batch_size=512 if device == 'cuda' else 128,
    convert_to_numpy=True,
    normalize_embeddings=True,
    show_progress_bar=True
).astype('float32')

# ======================================
# 🔹 FAISS 인덱스 구축
# ======================================
dim = embeddings.shape[1]
index = faiss.IndexFlatIP(dim)
index.add(embeddings)
print("✅ FAISS index size:", index.ntotal)

# ======================================
# 🔹 이미지 URL 메타정보 결합
# ======================================
# BOOK_IMAGE_NM → COVER_URL 로 이름 변경
df_books_meta = (
    df_books_contents
    .drop_duplicates(subset=['ISBN_THIRTEEN_NO'])
    .reset_index(drop=True)
)[['ISBN_THIRTEEN_NO', 'BOOK_IMAGE_NM', 'PBLICTE_DE']]

df_books_meta.rename(columns={'BOOK_IMAGE_NM': 'COVER_URL'}, inplace=True)

# ISBN 기준 merge
df_books = pd.merge(df_books, df_books_meta, on='ISBN_THIRTEEN_NO', how='left')

# 결측 이미지 처리
df_books['COVER_URL'] = df_books['COVER_URL'].replace('', np.nan).fillna('No Image')

# ======================================
# 🔹 파일 저장
# ======================================
save_path = "/content/drive/MyDrive/recommend_system/"

faiss.write_index(index, save_path + "book_faiss.index")
np.save(save_path + "book_embeddings.npy", embeddings)
df_books.to_pickle(save_path + "book_meta.pkl")

print("✅ 저장 완료!")
print(f"저장된 행 수: {len(df_books)}")
print(df_books[['BOOK_TITLE_NM', 'AUTHR_NM', 'COVER_URL']].head(3))
```
